```python
text=df['im getting on borderlands and i will murder you all ,'][5]
```

```python
import re

cleaned = []
for text in df['im getting on borderlands and i will murder you all ,']:
    text = str(text)  # Convert to string to avoid TypeError
    cleaned_text = re.sub(r'[^\w\s]', '', text)  # Remove special characters and punctuation
    cleaned_text = re.sub(r'\s+', ' ', cleaned_text)  # Replace multiple spaces with a single space
    cleaned_data = cleaned_text.strip()  # Remove leading/trailing whitespace
    cleaned.append(cleaned_data)
```

```python
print(cleaned_data)
```

```python
from nltk.tokenize import word_tokenize, sent_tokenize
from nltk.corpus import stopwords
tokens = [word_tokenize(x) for x in cleaned]
stop = set(stopwords.words('english'))  # Fetching the list of English stopwords
stpktn = []  # List to store stopword-removed tokens
for k in range(len(df['im getting on borderlands and i will murder you all ,'])):  # Loop through the tokenized text
    p = [i for i in tokens[k] if i not in stop]  # Filter out tokens that are in the stopword list
    stpktn.append(p) 
```

```python
import nltk
nltk.download('punkt')
nltk.download('punkt_tab')
```

```python
stpktn
```

```python
from nltk.stem import PorterStemmer
ps=PorterStemmer()
```

```python
for message in stpktn:
  [ps.stem(word) for word in message]
```

```python
stemmed_data=[]
for message in stpktn:
  stm=[ps.stem(word) for word in message]
  stemmed_data.append(stm)
```
